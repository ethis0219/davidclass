## 用selenium自動換頁爬取agoda飯店資訊
- 須用Firefox瀏覽器，並且下載selenium IDE https://addons.mozilla.org/zh-tw/firefox/addon/selenium-ide/

```python
from selenium import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

browser = webdriver.Firefox()    #開啟Firefox
browser.get('http://www.agoda.com/zh-tw/pages/agoda/default/DestinationSearchResult.aspx?asq=KI3WRnLKQeR4AGgsX7jigQrkOmaKhmogFnewFEcYpxHsH9%2fK3unCWZU9bCI07YkOM%2f3tnnz0wFLkQ9f0U%2fKZhXHrn0n6xgfO1jrV1TV%2bVbanNu7R5Z%2bL%2bAn1S%2foSEYuycyMSlq8K0cTE%2bmhPZLP%2b6BubLWWbHL40Lq76hhypS486Zl3pHAMTZy02KZ2Wb8gN&tick=636003876914')
soup = BeautifulSoup(browser.page_source)    # brower.page_source -> 取得網頁原始碼
while len(soup.select('.btn-right')) > 0:    # .btn-right為"下一頁"的標籤
    for ele in soup.select('.hotel-info h3'):    # 篩選飯店名稱
        print ele.text
    browser.find_element_by_id("paginationNext").click()    # 自動點擊下一頁
    soup = BeautifulSoup(browser.page_source)
browser.close()
```
